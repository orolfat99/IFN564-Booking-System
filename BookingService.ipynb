```python
from time import time
from random import random
import matplotlib.pyplot as plt
import numpy as np

repeats = 5
start = 1000
finish = 100000
step = 10000
algorithms =[AddCustomer,DeleteCustomer,RescheduleMovie,Screening]
colours =['red','blue','orange','purple']
fig, ax = plt.subplots(figsize=(10,10))

ax.set(xlabel='input_size', ylabel='Run time (sec)',title='Efficiency test')

index=0

for algorithm in algorithms:

        x=[]
        y=[]

        for input_size in range(start,finish+step,step):

            ys=[]
            for repeat in range(1,repeats+1):
                start_time = time()

                #generate the database
                data=[]
                for z in range(input_size):
                    data.append(str(random()))
#                     data.append(str(random()))
#                     print('Data:',data)

                #method to measure
                if algorithm == AddCustomer:
                    algorithm(data,data,data,data,data)
                elif algorithm == DeleteCustomer:
                    algorithm(data,data)
                elif algorithm == sale:
                    algorithm(data,data,data,data,data,data,data,data)
                elif algorithm == RescheduleMovie:
                    algorithm()
                elif algorithm == Screening:
                    algorithm(data,data,data,data)
                    
                    
                    

                elapsed_time = time() - start_time
#                 print(algorithm.__name__,input_size,repeat,elapsed_time,sep='\t')

                ys.append(elapsed_time)

            ax.scatter([input_size]*repeat,ys,c=colours[index],marker='.')
            
            y.append(sum(ys) / repeats)
            x.append(input_size)

        ax.plot(x, y,label=algorithm.__name__,c=colours[index])
        index+=1

plt.legend()
plt.show()
```
